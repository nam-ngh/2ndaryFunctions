```python
import pandas as pd
from src.pandas import compute_2D
from src.visualise import plot_2D
pd.set_option('display.max_columns', None)
data = pd.read_csv('data/melb_data.csv')
```

```python
data['Month'] = data['Date'].str[-4:] + '-' + data['Date'].str[-7:-5]
data.head()
```

1. FIND THE NUMBER OF LISTINGS FOR EACH REGION BY NUMBER OF ROOMS AVAILABLE

```python
region_room_count = compute_2D(df_all=data, d1='Regionname', d2='Rooms', total=False)
region_room_count
```

2. FIND THE NUMBER OF UNIQUE SELLERS EACH MONTH BY REGION AND PLOT

```python
sellers_count = compute_2D(df_all=data, d1='Month', d2='Regionname', d3='SellerG', method='nunique')
sellers_count
```

```python
fig = plot_2D(sellers_count)
fig.show()
```

3. CALCULATE MEAN PRICES FOR EACH SUBURB BY TIME, AND PLOT PRICE FOR ALBERT PARK, AIRPORT WEST, AND ESSENDON

```python
mean_price = compute_2D(df_all=data, d1='Month', d2='Suburb', d3='Price', method='mean')
mean_price
```
